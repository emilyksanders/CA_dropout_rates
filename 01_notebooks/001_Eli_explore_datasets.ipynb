# Read in Suspension Data

```python
# URL of the .txt file
url = 'https://www3.cde.ca.gov/demo-downloads/discipline/suspension23.txt'

# Read in the url with the delimiter and encoding to deal with special characters
df = pd.read_csv(url, delimiter='\t', encoding='ISO-8859-1')

df
```

```python
# Filter to only look at districts and total reporting category
df = df.loc[(df['AggregateLevel'] == 'D') & (df['ReportingCategory'] == 'TA') & (df['CharterYN'] == 'All')]
df
```

```python
df.drop(columns = ['AggregateLevel', 'SchoolCode', 'SchoolName', 'ReportingCategory', 'CharterYN'], inplace = True)
```

```python
df
```

```python
# Defining a function to grab all the years for us and concat them
def fetch_and_filter_suspension_data(year, existing_df=None):
    year_str = str(year)
    # Construct the URL based on the given year
    url = f'https://www3.cde.ca.gov/demo-downloads/discipline/suspension{year_str[-2:]}.txt'
    
    # Read the data from the URL with the specified delimiter and encoding
    df = pd.read_csv(url, delimiter='\t', encoding='ISO-8859-1')
    
    # Filter the DataFrame
    df = df.loc[(df['AggregateLevel'] == 'D') & (df['ReportingCategory'] == 'TA') & (df['CharterYN'] == 'All')]
    
    # Drop unnecessary columns
    df.drop(columns=['AggregateLevel', 'SchoolCode', 'SchoolName', 'ReportingCategory', 'CharterYN'], inplace=True)
    
    # Drop any columns that are unnamed
    df.drop(columns=[col for col in df.columns if 'Unnamed' in col], inplace=True)
    
    # Concatenate with the existing DataFrame if provided
    if existing_df is not None:
        df = pd.concat([existing_df, df], ignore_index=True)
    
    return df
```

```python
combined_df = None

# Add data for multiple years
for year in range(2012,2024):
    combined_df = fetch_and_filter_suspension_data(year, existing_df=combined_df)
    
# Combine the 'AcademicYear' columns by filling missing values in one with the other
if 'ï»¿AcademicYear' in combined_df.columns and 'AcademicYear' in combined_df.columns:
    combined_df['AcademicYear'] = combined_df['ï»¿AcademicYear'].fillna(combined_df['AcademicYear'])
    
# Combine the 'Cumulative Enrollment' columns similarly
if 'Cumulative Enrollment' in combined_df.columns and 'CumulativeEnrollment' in combined_df.columns:
    combined_df['Cumulative Enrollment'] = combined_df['Cumulative Enrollment'].fillna(combined_df['CumulativeEnrollment'])
# Drop the original columns that were combined
combined_df.drop(columns=['ï»¿AcademicYear', 'CumulativeEnrollment'], inplace=True, errors='ignore')
```

# Homeless Enrollment

```python
# URL of the .txt file
url = 'https://www3.cde.ca.gov/demo-downloads/homeless/hse2223.txt'

# Read in the url with the delimiter and encoding to deal with special characters
df = pd.read_csv(url, delimiter='\t', encoding='ISO-8859-1')

df
```

```python
df = df.loc[(df['Aggregate Level'] == 'D') & (df['Reporting Category'] == 'TA')]
df
```

```python
# Defining a function to grab all the years for us and concat them
def fetch_and_filter_homeless_data(start_year, end_year=None, existing_df=None):
    if end_year is None:
        end_year = start_year + 1
    year_str = f'{str(start_year)[-2:]}{str(end_year)[-2:]}'
    # Construct the URL based on the given year
    url = f'https://www3.cde.ca.gov/demo-downloads/homeless/hse{year_str}.txt'
    
    # Read the data from the URL with the specified delimiter and encoding
    df = pd.read_csv(url, delimiter='\t', encoding='ISO-8859-1')
    
    # Filter the DataFrame
    df = df.loc[(df['Aggregate Level'] == 'D') & (df['Reporting Category'] == 'TA')]
    
    # Drop unnecessary columns
    df.drop(columns=['Aggregate Level', 'School Code', 'School Name'], inplace=True)
    
    # Drop any columns that are unnamed
    df.drop(columns=[col for col in df.columns if 'Unnamed' in col], inplace=True)
    
    # Concatenate with the existing DataFrame if provided
    if existing_df is not None:
        df = pd.concat([existing_df, df], ignore_index=True)
    
    return df
```

```python
combined_df = None

# Add data for multiple years
for year in range(2019,2023):
    combined_df = fetch_and_filter_homeless_data(year, end_year=None, existing_df=combined_df)
```
